## Array Operations

11. Create two arrays: [1,2,3,4,5] and [6,7,8,9,10]. Perform element-wise addition, subtraction, multiplication, and division.

```python
arr1= np.array([1, 2, 3, 4, 5])
arr2= np.array([6, 7, 8, 9, 10])

print(arr1+arr2)
print(arr1-arr2)
print(arr1*arr2)
print(arr1/arr2)
```

12. Compute the square of every element in an array.

```python
square_1 = arr1 ** 2
square_2 = arr2 ** 2
print(square_1)
print(square_2)
```

13. Compute the sum of all elements in a NumPy array.

```python
print(arr1.sum())
print(arr2.sum())
```

14.Compute the mean, median, and standard deviation of a random array of size 20.

```python
arr = np.random.rand(20)
arr
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
```

15. Normalize an array (subtract mean and divide by standard deviation).

```python
mean = np.mean(arr)
std = np.std(arr)
Stand = (arr- mean) / std
print(Stand)
print(np.round(Stand, 2))
```

## Reshaping and Stacking

16. Create a 1D array of numbers from 1 to 12. Reshape it into a 3x4 matrix.

17. Flatten a 2D array into a 1D array.

18.Stack two arrays vertically.

```python
arr1=np.array([1,5,6,8,9])
arr2=np.array([7,9,11,12,15])
arrstack = np.vstack((arr1,arr2))
arrstack
```

19. Stack two arrays horizontally.

```python
arr1=np.array([1,5,6,8,9])
arr2=np.array([7,9,11,12,15])
arrstack = np.hstack((arr1,arr2))
arrstack
```

20. Split a 1D array of size 10 into two equal parts.

## Linear Algebra with NumPy

26. Create a 2x2 matrix and compute its determinant.

```python
matrix= np.arange(1,5).reshape(2,2)
print(matrix)
print(round(np.linalg.det(matrix)))
```

27. Compute the inverse of a 3x3 matrix.

```python
matrix=np.random.rand(3,3)
print(matrix)
print(np.linalg.inv(matrix))
```

28. Perform matrix multiplication on two compatible matrices.

```python
MatA = np.random.rand(3,2)
MatB = np.random.rand(2,3)
print(MatA)
print(MatB)
print(MatA @ MatB)
```

29. Compute the eigenvalues and eigenvectors of a 2x2 matrix.

```python
matrix = np.array(
   [[4, 1],
    [2, 3] ])
eigenvalues, eigenvectors = np.linalg.eig(matrix)
print(eigenvalues)
print(eigenvectors)
```

30. Solve the system of linear equations: 2x + y = 5, x - y = 1.

```python
X = np.array([
    [2, 1],
    [1, -1]
])

Y = np.array([5, 1])

solution = np.linalg.solve(X, Y)

print(solution)
```
